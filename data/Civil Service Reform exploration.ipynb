```python
df = pd.read_excel("raw/Government_Major_Projects_Portfolio_data__September_2013__XLS_.xlsx", 0)
```

```python
df = df.set_index("Project name").T
df = df.rename(columns={
  df.columns[1]:"rating",
  df.columns[2]:"description",
  df.columns[3]:"department_commentary",
  df.columns[4]:"date_start",
  df.columns[5]:"date_end",
  df.columns[6]:"department_narrative_on_schedule",
  df.columns[7]: "2013_cash_budget",
  df.columns[8]: "2013_cash_forecast",
  df.columns[9]: "2013_cash_variance",
  df.columns[10]:"2013_percent_variance",
  df.columns[11]:"total_lifebudget",
  df.columns[12]:"department_narrative_on_budget_variance",
  df.columns[13]:"department_narrative_on_budget_life",
})
df.head()
```

```python
numeric_columns = ["2013_cash_budget", "2013_cash_forecast", "2013_cash_variance", "2013_percent_variance", "total_lifebudget"]
for column in numeric_columns:
    df[column] = df[column].map(lambda x: float("nan") if isinstance(x, basestring) else x)
```

```python
def rating_to_colour(rating):
    colour = "#000000"
    if rating == "Green":
        colour = "#00ff00"
    elif rating == "Amber/Green" or rating == "Green/Amber":
        colour = "#000000"
    elif rating == "Amber":
        colour = "#ffff00"
    elif rating == "Amber/Red" or rating == "Red/Amber":
        colour = "#000000"
    elif rating == "Red":
        colour = "#ff0000"
    return colour
df["colour"] = df.apply(lambda row: rating_to_colour(row["rating"]), axis=1)
```

```python
df.plot(x="2013_cash_budget", y="2013_cash_forecast", kind="scatter", c=df["colour"])
```

```python
def rating_to_number(rating):
    number = float("nan")
    if rating == "Green":
        number = 4
    elif rating == "Amber/Green" or rating == "Green/Amber":
        number = 3
    elif rating == "Amber":
        number = 2
    elif rating == "Amber/Red" or rating == "Red/Amber":
        number = 1
    elif rating == "Red":
        number = 0
    return number
df["numeric_rating"] = df.apply(lambda row: rating_to_number(row["rating"]), axis=1)
```

```python
df.plot(x="2013_cash_budget", y="numeric_rating", kind="scatter")
```
