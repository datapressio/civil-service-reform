```python
import itertools
```

```python
[x for x in itertools.groupby(latestByDeptAndCustomerType.iterrows(), lambda ((d,c),v): d)]
```

```python
def format_value(value):
    return (value if not np.isnan(value) else None)
```

```python
def format_row(data):
    return {
      "digital_volume": format_value(data["Latest digital volume"]),
      "total_volume":  format_value(data["Latest volume"]),
      "digital_proportion": format_value(data["Latest digital proportion"])
    }
```

```python
def format_dept(iterator):
    return dict((k2, format_row(v)) for ((k1, k2), v) in iterator)
```

```python
formatted = dict((dept, format_dept(it)) for (dept, it) in itertools.groupby(latestByDeptAndCustomerType.iterrows(), lambda ((d, c), v): d))
    
```

```python
with open("digital_transactions_departmental_breakdown.json", "w") as f:
    json.dump(formatted, f, indent=True, allow_nan=False)
```
