```python
f = 'raw/Government_Major_Projects_Portfolio_data__September_2013__XLS_.xlsx'
import pandas
```

```python
df=pandas.read_excel(f,0)
df = df.set_index("Project name").T
df = df.rename(columns={
  df.columns[1]:"rating",
  df.columns[2]:"description",
  df.columns[3]:"department_commentary",
  df.columns[4]:"date_start",
  df.columns[5]:"date_end",
  df.columns[6]:"department_narrative_on_schedule",
  df.columns[7]: "2013_cash_budget",
  df.columns[8]: "2013_cash_forecast",
  df.columns[9]: "2013_cash_variance",
  df.columns[10]:"2013_percent_variance",
  df.columns[11]:"total_lifebudget",
  df.columns[12]:"department_narrative_on_budget_variance",
  df.columns[13]:"department_narrative_on_budget_life",
})
df.head()
```

```python
df.to_csv("major_projects.csv",encoding="utf-8")
```
