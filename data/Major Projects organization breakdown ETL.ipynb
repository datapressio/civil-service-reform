```python
def format_departments(data):
    departments = data["Department"].unique()
    return dict(format_department(department, data) for department in departments)
```

```python
def format_ratings(data):
    ratings = data["rating"].unique()
    return dict(format_rating(rating, data) for rating in ratings)
```

```python
def format_rating(rating, data):
    projects_data = data[data["rating"] == rating]
    summary = format_summary(projects_data)
    projects = format_projects(projects_data)
    value = { "summary": summary, "projects": projects}
    return (rating, value)
```

```python
def format_department(department, data):
    projects_data = data[data["Department"] == department]
    summary = format_summary(projects_data)
    projects = format_projects(projects_data)
    value = { "summary": summary, "projects" : projects }
    return (department, value)
```

```python
def format_projects(data):
    return dict(format_project(name, project) for name, project in data.iterrows())
```

```python
def format_summary(data):
    total_budget = data["2013_cash_budget"][data["2013_cash_budget"].map(np.isreal)].sum()
    total_forecast = data["2013_cash_forecast"][data["2013_cash_forecast"].map(np.isreal)].sum()
    total_variance = total_forecast - total_budget
    percent_variance = total_variance / total_budget * 100 if total_budget > 0 else 0
    total_lifebudget = data["total_lifebudget"][data["total_lifebudget"].map(np.isreal)].sum()
    under_budget = data[data["2013_cash_variance"] < 0]
    over_budget = data[data["2013_cash_variance"] > 0]
    values = {
        "n" : len(data),
        "n_under_budget": len(under_budget),
        "n_over_budget": len(over_budget),
        "2013_cash_budget": int(math.floor(total_budget)),
        "2013_cash_forecast": int(math.floor(total_forecast)),
        "2013_cash_variance": int(math.floor(total_variance)),
        "2013_percent_variance": percent_variance,
        "total_lifebudget": int(math.floor(total_lifebudget)),
        "n_red": len(data[data["rating"] == "Red"]),
        "n_amber_red" : len(data[data["rating"] == "Amber/Red"]),
        "n_amber" : len(data[data["rating"] == "Amber"]),
        "n_amber_green" : len(data[data["rating"] == "Amber/Green"]),
        "n_green" : len(data[data["rating"] == "Green"])        
    }
    return values
```

```python
def format_project(name, row):
    values = dict((k, format_value(k, row[k])) for k in row.keys())
    return (name, values)
```

```python
def millis_since_epoch(time):
    epoch = datetime.datetime.fromtimestamp(0)
    delta = (time - epoch).total_seconds()
    return int(delta) * 1000
```

```python
def format_value(key, value):
    if isinstance(value, float) and np.isnan(value):
        value = None
    elif isinstance(value, basestring) and re.compile("^(Data exempt|Exempt|Reset|n/a|TBC)").match(value):
        value = None
    elif key in ["date_start", "date_end"]:
        value = millis_since_epoch(value)
    elif key in ["2013_cash_budget", "2013_cash_forecast", "2013_cash_variance", "total_lifebudget"]:
        value = int(math.floor(value))
    elif key == "2013_percent_variance":
        value = value * 100
    return value
```

```python
def disambiguate_departments(df):
    ddf = df.copy()
    ddf.ix[ddf["Department"] == "MOJ", "Department"] = "MoJ"
    ddf.ix[ddf["Department"] == "DOH", "Department"] = "DoH"
    return ddf
```
